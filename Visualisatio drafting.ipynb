```python
# read dataset
file_handler = open('flights_per_month_top_airports.pkl','rb')
flights_per_month = pickle.load(file_handler)
file_handler.close()
```

```python
# normalize traffic vs 2000 January as reference
tst = flights_per_month.T.astype('float32')
tst = tst.div(tst['2000-01'], axis='index')
flights_per_month_normalized = tst.T
#flights_per_month_normalized.head()
```
